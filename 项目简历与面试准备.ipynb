### Q1：为什么采用多 Agent 架构而不是单 Agent？

**A：**  
单 Agent 需要在一个 Prompt 中处理全部任务，Prompt 过长且容易产生幻觉。拆分为 5 个专业 Agent 的好处：

1. **职责单一**：每个 Agent 只关注一个领域（景点/天气/酒店/美食/规划），System Prompt 更精准，工具调用更准确
2. **并行加速**：前 4 个数据采集 Agent 相互独立，可以通过 `asyncio.gather` 全并行执行，耗时从 ~30s 降至 ~8s
3. **可扩展性**：新增 Agent（如交通 Agent）只需新增类 + 注册到编排器，不影响现有流程
4. **容错隔离**：`asyncio.gather(return_exceptions=True)` 使单个 Agent 失败不会导致整体崩溃，配合 Fallback 备用方案

---

### Q2：5 个 Agent 之间是如何协作的？

**A：** 采用**编排者模式（Orchestrator Pattern）**。`MultiAgentTripPlanner` 类作为编排层，实现**分阶段并行-串行混合执行**：

**第一阶段（并行）**：4 个数据采集 Agent 全并行
```python
all_tasks = [
    *[asyncio.to_thread(self.attraction_agent.run, q) for q in attraction_queries],  # 每个城市一个任务
    *[asyncio.to_thread(self.weather_agent.run, q) for q in weather_queries],
    asyncio.to_thread(self.hotel_agent.run, hotel_query),
    asyncio.to_thread(self.food_agent.run, food_query),
]
all_results = await asyncio.gather(*all_tasks, return_exceptions=True)
```

**第二阶段（串行）**：Planner Agent 接收全部数据，整合生成结构化 JSON
- Planner Agent **不带任何工具**，只负责信息聚合和结构化输出，这是职责分离的关键设计

**第三阶段（后处理）**：坐标修正 → 天气预警 → 开放时间 → 缓存写入

---

### Q3：为什么使用 `asyncio.to_thread()` 而不是直接 `await`？

**A：**  
`SimpleAgent.run()` 是**同步阻塞方法**（内部调用 LLM + MCP 工具链），不是 `async` 协程。直接在事件循环中调用会阻塞整个 FastAPI 线程。

`asyncio.to_thread()` 将同步函数放入**线程池**执行，使得多个 Agent 可以在不同线程中并发运行，同时不阻塞 FastAPI 的事件循环处理其他请求。

```python
# ✅ 正确：线程池并行
await asyncio.gather(
    asyncio.to_thread(agent1.run, q1),  # 线程A
    asyncio.to_thread(agent2.run, q2),  # 线程B
)

# ❌ 错误：顺序阻塞
result1 = agent1.run(q1)   # 阻塞等待
result2 = agent2.run(q2)   # 再阻塞等待
```

---

### Q4：如果增加一个Agent，怎么扩展？

**A：**
1. 在文件中新增 Agent 的 System Prompt 常量（如 `TRANSPORT_AGENT_PROMPT`）
2. 在 `__init__` 中创建 `self.transport_agent = SimpleAgent(...)` 并注册 MCP 工具
3. 在 `plan_trip_stream()` 的 `all_tasks` 列表中添加新的 `asyncio.to_thread(...)` 任务
4. 在结果收集阶段提取新 Agent 的结果，传递给 Planner Agent
5. 更新 Planner Agent 的 Prompt，告知需要整合新数据源

扩展过程不需要修改已有 Agent，体现了**开闭原则**。

---

### Q5：多城市路线如何分配 Agent 任务？

**A：**  
多城市模式（如"北京→西安→成都"，6天）：

1. **天数分配**：`days_per_city = total_days / len(cities)`，即每城市 2 天
2. **任务拆分**：为**每个城市**各生成一个景点搜索 query 和天气 query，酒店/美食聚合所有城市
3. **结果标注**：用 `【北京景点信息】`、`【西安景点信息】` 等标签区分不同城市数据
4. **Planner 指令**：在 Prompt 中注明"前 2 天安排北京，第 3-4 天安排西安，第 5-6 天安排成都，城市间安排交通衔接"

```python
for i, city in enumerate(city_list):
    start_day = i * days_per_city + 1
    end_day = (i + 1) * days_per_city
    planner_query += f"\n- 第{start_day}-{end_day}天安排{city}的行程"
```

## 四、面试 Q&A — MCP 协议（重要技术点）

---

### Q1：什么是 MCP 协议？你是如何使用它的？

**A：**  
MCP（Model Context Protocol）是 Anthropic 提出的开放协议，用于标准化 LLM 应用与外部工具/数据源的交互，类似于 AI 系统的「USB-C 接口」。

在本项目中：
- 使用 `MCPTool` 类通过 **stdio 传输**方式启动 `amap-mcp-server`（高德地图 MCP Server）
- MCP Server 暴露 16 个工具（`maps_text_search`、`maps_weather`、路线规划等）
- Agent 通过 MCP 协议自动发现并调用这些工具，无需手写 HTTP 请求

```python
self.amap_tool = MCPTool(
    name="amap",
    server_command=["uvx", "amap-mcp-server"],
    env={"AMAP_MAPS_API_KEY": settings.amap_api_key},
    auto_expand=True
)
# 手动展开子工具后注册到各 Agent
expanded_tools = self.amap_tool.get_expanded_tools()
for t in expanded_tools:
    self.attraction_agent.add_tool(t)
    self.weather_agent.add_tool(t)
    self.hotel_agent.add_tool(t)
    self.food_agent.add_tool(t)
```

---

### Q2：MCP 的 stdio 传输方式是什么意思？

**A：** MCP 支持两种传输方式：

| 传输方式 | 原理 | 适用场景 |
|----------|------|----------|
| **stdio** | 客户端以子进程方式启动 MCP Server，通过标准输入/输出进行 JSON-RPC 通信 | 本地工具集成 |
| **HTTP/SSE** | 通过 HTTP 请求 + Server-Sent Events 通信 | 远程服务 |

本项目用 stdio 方式，`server_command=["uvx", "amap-mcp-server"]`，即通过 `uvx`（uv 的工具执行器）直接运行高德地图 MCP Server 包。优点是无需独立部署服务，Agent 启动时自动拉起 MCP Server 子进程。

---

### Q3：MCPTool 默认不展开子工具的坑是怎么发现并解决的？

**A：**  
**现象**：Agent 不调用任何工具，直接用 LLM 编造景点数据。  
**排查**：发现 `attraction_agent.list_tools()` 只返回 1 个工具（`amap` 聚合工具），而非 16 个子工具。  
**根因**：`MCPTool` 继承自 `Tool`，`expandable` 属性默认为 `False`，`add_tool()` 不会自动展开。  
**解决**：调用 `amap_tool.get_expanded_tools()` 手动展开，将 16 个子工具逐一 `add_tool()` 到每个 Agent。

> **面试包装**：这体现了对框架源码的深入理解。当 Agent 行为异常时，我首先检查工具注册情况而非怀疑 LLM 能力，通过 `list_tools()` 定位到工具未展开，然后阅读 MCPTool 源码发现 expandable 默认值机制。

---

### Q4：MCP 与 Function Calling 有什么区别？

**A：**

| 维度 | Function Calling | MCP |
|------|-----------------|-----|
| **标准** | 各厂商私有（OpenAI、Claude 格式不同） | 统一开放标准 |
| **工具发现** | 开发者在代码中定义工具 schema | Server 自动暴露工具列表 |
| **传输方式** | 嵌入在 LLM API 请求中 | 独立的 JSON-RPC 通信 |
| **适用范围** | 单次工具调用 | 持久连接，支持多轮工具调用 |

MCP 的核心价值在于**标准化**：同一个 MCP Server（如高德地图）可以被任何框架（LangChain、HelloAgents、Claude）直接接入，无需适配。

## 五、面试 Q&A — SSE 流式推送与缓存（进阶亮点）

---

### Q1：为什么使用 SSE 而不是 WebSocket？

**A：**  
本场景是**服务端单向推送进度**（Agent 执行百分比 → 前端进度条），不需要双向通信。

| 维度 | SSE | WebSocket |
|------|-----|-----------|
| **方向** | 单向（服务端→客户端） | 双向 |
| **协议** | 基于 HTTP/1.1 | 独立协议（ws://） |
| **重连** | 浏览器自动重连 | 需手动实现 |
| **复杂度** | 低（标准 HTTP 流） | 高（连接管理/心跳） |
| **Nginx 支持** | 简单（proxy_buffering off） | 需额外配置 upgrade |

SSE 更轻量，且 FastAPI 原生支持 `StreamingResponse`，无需额外依赖。

---

### Q2：SSE 的后端实现原理？

**A：**  

```python
@router.post("/plan/stream")
async def plan_trip_stream(request: TripRequest):
    async def event_generator():
        async for event in planner.plan_trip_stream(request):
            yield f"data: {json.dumps(event, ensure_ascii=False)}\n\n"
    
    return StreamingResponse(
        event_generator(),
        media_type="text/event-stream",
        headers={
            "Cache-Control": "no-cache",
            "Connection": "keep-alive",
            "X-Accel-Buffering": "no",   # 告知 Nginx 不缓冲
        }
    )
```

**事件流格式（SSE 规范）**：
```
data: {"type": "progress", "percent": 15, "message": "🔄 并行搜索景点、天气..."}\n\n
data: {"type": "progress", "percent": 65, "message": "📋 整合信息，生成行程..."}\n\n
data: {"type": "done", "data": {...完整TripPlan...}}\n\n
```

---

### Q3：前端如何解析 SSE 流？

**A：** 使用 Fetch API 读取 `ReadableStream`，而非 `EventSource`（因为需要 POST + JSON body）：

```typescript
const response = await fetch(url, { method: 'POST', body: JSON.stringify(data) })
const reader = response.body!.getReader()
const decoder = new TextDecoder()
let buffer = ''

while (true) {
  const { done, value } = await reader.read()
  if (done) break
  buffer += decoder.decode(value, { stream: true })
  
  // 按行分割，解析 "data: {...}" 格式
  const lines = buffer.split('\n')
  buffer = lines.pop()!  // 保留不完整行
  for (const line of lines) {
    if (line.startsWith('data: ')) {
      const event = JSON.parse(line.slice(6))
      if (event.type === 'progress') onProgress(event)
      if (event.type === 'done') onComplete(event.data)
    }
  }
}
```

> **要点**：使用 `buffer` 处理不完整行（TCP 分包问题），这是流式解析的关键。

---

### Q4：缓存系统是如何设计的？

**A：** 自实现的 **TTL 内存缓存**（无外部依赖）：

```python
class TTLCache:
    def __init__(self, ttl_seconds=3600):  # 1小时过期
        self._store: Dict[str, Dict] = {}
        self._ttl = ttl_seconds
    
    def get(self, key):
        item = self._store.get(key)
        if not item: return None
        if time.time() - item["ts"] > self._ttl:
            del self._store[key]  # 惰性淘汰
            return None
        return item["data"]
```

**缓存 Key 生成**：对请求参数做 MD5 哈希，确保相同查询条件命中同一 Key：
```python
def make_trip_cache_key(city, cities, start_date, end_date, preferences, ...):
    raw = json.dumps({
        "city": city, "cities": sorted(cities),
        "start_date": start_date, "preferences": sorted(preferences), ...
    }, sort_keys=True)
    return "trip:" + hashlib.md5(raw.encode()).hexdigest()[:12]
```

**设计选择**：为什么不用 Redis？
1. 项目规模单机部署，无需分布式缓存
2. 减少运维复杂度（无需部署 Redis 容器）
3. 旅行规划结果较大（~50KB JSON），内存缓存读写零延迟
4. 后续扩展到 Redis 只需替换 `TTLCache` 实现，接口不变

---

### Q5：缓存 Key 使用 MD5 有安全问题吗？

**A：** 没有。此处 MD5 用于**生成唯一标识**（类似 etag），不用于密码加密或签名验证。MD5 的碰撞风险在此场景下可以忽略不计（2^64 次才有 50% 概率碰撞），且取前 12 位十六进制字符已足够区分不同请求。

### Q1：为什么选择 FastAPI？

**A：**

### Q2：LLM 实例是如何管理的？

**A：** 使用**全局单例模式**。`get_llm()` 函数首次调用时创建 `HelloAgentsLLM` 实例并缓存到模块级变量 `_llm_instance`。

```python
_llm_instance = None

def get_llm() -> HelloAgentsLLM:
    global _llm_instance
    if _llm_instance is None:
        _llm_instance = HelloAgentsLLM(provider="custom")
    return _llm_instance
```

**关键细节**：显式指定 `provider="custom"` 是为了防止系统环境中已有的 `DASHSCOPE_API_KEY` 让 HelloAgents 自动选择 Qwen provider，导致 API Key 发送到错误端点返回 401。

### Q3：Pydantic V2 的 field_validator 如何处理 LLM 返回类型不一致的问题？

**A：** LLM 返回的 JSON 字段类型不可控，使用 `mode='before'` 验证器在 Pydantic 类型转换前拦截并预处理：

```python
class DayPlan(BaseModel):
    accommodation: Optional[str] = ""

    @field_validator('accommodation', mode='before')
    @classmethod
    def parse_accommodation(cls, v):
        if v is None: return ""
        if isinstance(v, dict): return v.get('name') or v.get('type') or str(v)
        return v

class WeatherInfo(BaseModel):
    day_temp: Union[int, str]

    @field_validator('day_temp', 'night_temp', mode='before')
    @classmethod
    def parse_temperature(cls, v):
        if isinstance(v, str):
            v = v.replace('°C', '').replace('℃', '').strip()
            try: return int(v)
            except ValueError: return 0
        return v
```

**`mode='before'` vs `mode='after'`**：`before` 在类型转换前执行（输入是原始 JSON 数据），`after` 在转换后执行（输入已是目标类型）。LLM 场景几乎都用 `before`。

### Q4：CORS 是怎么配置的？

## 七、面试 Q&A — LLM 响应解析与容错

---

### Q1：LLM 返回数据不可控，你是怎么处理的？

**A：** 三层策略：

**第一层 — Prompt 约束**：在 Planner Agent 的 System Prompt 中提供完整 JSON Schema 示例，明确每个字段的数据类型和含义，末尾强调 `**只返回JSON，不要有任何说明文字**`

**第二层 — 多格式 JSON 提取**：

```python
def _parse_response(self, response: str, request: TripRequest) -> TripPlan:
    # 策略1: ```json 代码块
    if "```json" in response:
        json_start = response.find("```json") + 7
        json_end = response.find("```", json_start)
        json_str = response[json_start:json_end].strip()
    # 策略2: 普通 ``` 代码块
    elif "```" in response:
        json_start = response.find("```") + 3
        json_end = response.find("```", json_start)
        json_str = response[json_start:json_end].strip()
    # 策略3: 直接定位大括号
    elif "{" in response and "}" in response:
        json_start = response.find("{")
        json_end = response.rfind("}") + 1
        json_str = response[json_start:json_end]
    
    data = json.loads(json_str)
    return TripPlan(**data)  # Pydantic 校验 + field_validator 容错
```

**第三层 — Fallback 备用方案**：`_create_fallback_plan()` 在任何解析失败时生成包含占位景点的基本行程（使用城市中心坐标缓存），确保用户不会看到错误页面。

### Q2：Fallback 方案具体是怎么做的？

**A：**

```python
def _create_fallback_plan(self, request: TripRequest) -> TripPlan:
    # 预置热门城市坐标
    city_centers = {
        "北京": {"longitude": 116.397428, "latitude": 39.90923},
        "上海": {"longitude": 121.473701, "latitude": 31.230416},
        "广州": {"longitude": 113.264385, "latitude": 23.12911},
        # ... 更多城市
    }
    
    # 为每天生成 2 个占位景点 + 三餐
    for day_idx in range(request.travel_days):
        day = DayPlan(
            date=start_date + timedelta(days=day_idx),
            day_index=day_idx + 1,
            attractions=[Attraction(name=f"{city}推荐景点{i+1}", ...)],
            meals=[
                Meal(type="breakfast", name=f"{city}特色早餐", ...),
                Meal(type="lunch", name=f"{city}特色午餐", ...),

```

**设计原则**：Fallback 宁可给出不精确的内容，也要保证**完整的数据结构**，使前端渲染不会报错。

---

### Q3：为什么 JSON 提取不直接用正则？

**A：** 实际也可以用正则（如 `re.search(r'\{.*\}', response, re.DOTALL)`），但有以下考量：
1. LLM 可能返回嵌套 JSON，贪婪匹配 `.*` 可能匹配到错误的 `}`
2. `rfind("}") + 1` 取最后一个 `}` 比正则更可靠（JSON 对象总是以最外层 `}` 结尾）
3. 优先用 `\`\`\`json` 标记定位比盲目正则更准确
4. 性能差异可忽略，清晰度更重要

## 思考：何时截断/滑动窗口优于 LLM 摘要？

| 场景 | 推荐策略 | 原因 |
|---|---|---|
| 超预算 **< 15%** | 截断 | 损失少，无需额外 API 调用 |
| 内容有**严格顺序依赖**（代码、日志） | 滑动窗口 | LLM 摘要易丢失细节导致语义错误 |
| **实时/低延迟**场景 | 截断/滑动窗口 | LLM 摘要增加 1-3s 延迟 |
| **高置信度结构化**内容（表格、JSON） | 截断 | 摘要会破坏结构 |
| 超预算 **> 40%** 且内容为自然语言 | LLM 摘要 | 截断损失过大，摘要 ROI 高 |
| **跨会话长期记忆** | LLM 摘要 | 需语义压缩，截断无意义 |

**核心判断原则**：压缩代价（延迟 + 费用）是否低于信息损失代价。

## 混合压缩策略设计



````python
"""ContextBuilder - GSSC流水线实现"""

from typing import Dict, Any, List, Optional, Tuple
from dataclasses import dataclass, field
from datetime import datetime
import tiktoken
import math
import re

from ..core.message import Message
from ..tools import MemoryTool, RAGTool

# ...existing code...

# ═══════════════════════════════════════════════════════════════════════════════
# 混合压缩策略
# ═══════════════════════════════════════════════════════════════════════════════

class CompressionMethod:
    """压缩方法枚举"""
    SKIP        = "skip"        # 无需压缩
    TRUNCATE    = "truncate"    # 尾部截断
    SLIDING     = "sliding"     # 滑动窗口（保留首尾）
    DEDUP       = "dedup"       # 去重
    LLM_SUMMARY = "llm_summary" # LLM 摘要


@dataclass
class CompressionDecision:
    """压缩决策结果"""
    method: str                    # 选用的方法
    overflow_ratio: float          # 超出比例
    content_type: str              # 内容类型
    reason: str                    # 决策原因
    estimated_savings: int         # 预估节省 tokens


@dataclass
class CompressionResult:
    """压缩结果"""
    compressed_text: str
    original_tokens: int
    final_tokens: int
    method_used: str
    compression_ratio: float       # final / original
    decisions: List[CompressionDecision] = field(default_factory=list)

    def __str__(self) -> str:
        saved = self.original_tokens - self.final_tokens
        return (
            f"[压缩结果] {self.method_used} | "
            f"{self.original_tokens} → {self.final_tokens} tokens "
            f"(节省 {saved}, 压缩率 {self.compression_ratio:.2f})"
        )


class HybridCompressor:
    """混合压缩器
    
    决策树：
    
                     ┌─ overflow < 15% ──→ TRUNCATE
                     │
    overflow > 0% ───┼─ 15% ≤ overflow < 40% ──→ DEDUP → SLIDING → (若仍超) TRUNCATE
                     │
                     └─ overflow ≥ 40% ─────────→ DEDUP → LLM_SUMMARY → (兜底) SLIDING+TRUNCATE
    
    内容类型修正：
    - 代码/JSON/日志  → 禁用 LLM_SUMMARY，改用 SLIDING
    - 结构化表格      → 优先 TRUNCATE（保留表头）
    - 自然语言段落    → 允许 LLM_SUMMARY
    """

    # 内容类型检测正则
    _CODE_PATTERN = re.compile(
        r'```|def |class |import |SELECT |INSERT |{.*?}|\[.*?\]',
        re.DOTALL
    )
    _JSON_PATTERN  = re.compile(r'^\s*[\[{]', re.MULTILINE)
    _TABLE_PATTERN = re.compile(r'\|.*\|.*\|')
    _LOG_PATTERN   = re.compile(
        r'\d{4}-\d{2}-\d{2}|\[INFO\]|\[ERROR\]|\[WARN\]|ERROR:|Traceback'
    )

    def __init__(
        self,
        llm=None,                           # HelloAgentsLLM 实例，可选
        llm_summary_min_overflow: float = 0.40,  # 启用 LLM 摘要的最小超出比例
        llm_summary_max_tokens: int = 500,       # LLM 摘要最大输出 tokens
        sliding_window_overlap: int = 50,         # 滑动窗口首尾重叠 tokens
        enable_dedup: bool = True,
    ):
        self.llm = llm
        self.llm_summary_min_overflow = llm_summary_min_overflow
        self.llm_summary_max_tokens = llm_summary_max_tokens
        self.sliding_window_overlap = sliding_window_overlap
        self.enable_dedup = enable_dedup

    # ── 公开入口 ──────────────────────────────────────────────────────────────

    def compress(self, context: str, budget: int) -> CompressionResult:
        """主压缩入口
        
        Args:
            context: 待压缩文本
            budget:  目标 token 预算
            
        Returns:
            CompressionResult
        """
        original_tokens = count_tokens(context)

        # 无需压缩
        if original_tokens <= budget:
            return CompressionResult(
                compressed_text=context,
                original_tokens=original_tokens,
                final_tokens=original_tokens,
                method_used=CompressionMethod.SKIP,
                compression_ratio=1.0
            )

        overflow_ratio = (original_tokens - budget) / original_tokens
        content_type   = self._detect_content_type(context)
        decisions: List[CompressionDecision] = []

        # 按章节分别压缩（保护结构）
        text = self._compress_by_section(
            context, budget, overflow_ratio, content_type, decisions
        )

        final_tokens = count_tokens(text)
        # 最终兜底：若仍超预算，强制截断
        if final_tokens > budget:
            text = self._truncate(text, budget)
            decisions.append(CompressionDecision(
                method=CompressionMethod.TRUNCATE,
                overflow_ratio=overflow_ratio,
                content_type=content_type,
                reason="兜底截断：前序策略未达预算",
                estimated_savings=final_tokens - count_tokens(text)
            ))
            final_tokens = count_tokens(text)

        method_summary = " → ".join(d.method for d in decisions) or CompressionMethod.SKIP
        return CompressionResult(
            compressed_text=text,
            original_tokens=original_tokens,
            final_tokens=final_tokens,
            method_used=method_summary,
            compression_ratio=final_tokens / max(original_tokens, 1),
            decisions=decisions
        )

    # ── 内容类型检测 ─────────────────────────────────────────────────────────

    def _detect_content_type(self, text: str) -> str:
        """检测主要内容类型"""
        sample = text[:2000]  # 取前2000字符做采样
        if self._JSON_PATTERN.search(sample) and text.count('{') > 3:
            return "json"
        if self._CODE_PATTERN.search(sample):
            return "code"
        if self._LOG_PATTERN.search(sample):
            return "log"
        if self._TABLE_PATTERN.search(sample):
            return "table"
        return "prose"  # 自然语言段落

    def _can_use_llm_summary(self, content_type: str, overflow_ratio: float) -> bool:
        """判断是否允许 LLM 摘要"""
        if self.llm is None:
            return False
        if content_type in ("code", "json", "log"):
            return False   # 结构化内容禁止 LLM 摘要
        return overflow_ratio >= self.llm_summary_min_overflow

    # ── 章节级压缩 ───────────────────────────────────────────────────────────

    def _compress_by_section(
        self,
        context: str,
        budget: int,
        overflow_ratio: float,
        content_type: str,
        decisions: List[CompressionDecision]
    ) -> str:
        """按章节优先级分配 token 预算，各章节独立压缩"""
        # 章节优先级（数字越大越优先保留）
        SECTION_PRIORITY = {
            "Role & Policies": 5,
            "Task":            5,   # 必须完整保留
            "Output":          5,   # 必须完整保留
            "State":           3,
            "Evidence":        2,
            "Context":         1,   # 最低优先级，最先压缩
        }

        section_pattern = re.compile(
            r'(?=\[(Role & Policies|Task|State|Evidence|Context|Output)\])',
            re.MULTILINE
        )
        parts = section_pattern.split(context)

        # 解析章节
        sections: List[Tuple[str, str, int]] = []  # (name, content, priority)
        for part in parts:
            if not part.strip():
                continue
            match = re.match(r'\[(.*?)\]', part)
            name = match.group(1) if match else "Unknown"
            priority = SECTION_PRIORITY.get(name, 2)
            sections.append((name, part, priority))

        # 按优先级从低到高排序，低优先级先压缩
        sections_sorted = sorted(sections, key=lambda x: x[2])

        compressed_sections: Dict[str, str] = {}
        current_total = sum(count_tokens(s[1]) for s in sections)

        for name, content, priority in sections_sorted:
            if current_total <= budget:
                break  # 已达预算，后续章节不压缩

            section_tokens = count_tokens(content)
            # 该章节需要压缩的量
            need_to_free = current_total - budget
            section_budget = max(section_tokens - need_to_free, section_tokens // 2)

            compressed = self._compress_section(
                name, content, section_budget,
                overflow_ratio, content_type, decisions
            )
            compressed_sections[name] = compressed
            saved = section_tokens - count_tokens(compressed)
            current_total -= saved

        # 重组：按原始顺序拼接
        result_parts = []
        for name, content, _ in sections:
            result_parts.append(compressed_sections.get(name, content))

        return "\n\n".join(result_parts)

    def _compress_section(
        self,
        section_name: str,
        content: str,
        budget: int,
        overflow_ratio: float,
        content_type: str,
        decisions: List[CompressionDecision]
    ) -> str:
        """对单个章节执行压缩流水线：DEDUP → 策略压缩"""
        text = content

        # Step 1: 去重（所有类型通用）
        if self.enable_dedup and count_tokens(text) > budget:
            before = count_tokens(text)
            text = self._dedup_lines(text)
            after = count_tokens(text)
            if before != after:
                decisions.append(CompressionDecision(
                    method=CompressionMethod.DEDUP,
                    overflow_ratio=overflow_ratio,
                    content_type=content_type,
                    reason=f"[{section_name}] 去除重复行",
                    estimated_savings=before - after
                ))

        if count_tokens(text) <= budget:
            return text

        # Step 2: 根据超出比例和内容类型选择策略
        if overflow_ratio < 0.15:
            # 轻微超出 → 直接截断
            text = self._truncate(text, budget)
            decisions.append(CompressionDecision(
                method=CompressionMethod.TRUNCATE,
                overflow_ratio=overflow_ratio,
                content_type=content_type,
                reason=f"[{section_name}] 轻微超出(<15%)，尾部截断",
                estimated_savings=count_tokens(content) - count_tokens(text)
            ))

        elif self._can_use_llm_summary(content_type, overflow_ratio):
            # 超出≥40% 且为自然语言 → LLM 摘要
            text = self._llm_summarize(text, budget, section_name)
            decisions.append(CompressionDecision(
                method=CompressionMethod.LLM_SUMMARY,
                overflow_ratio=overflow_ratio,
                content_type=content_type,
                reason=f"[{section_name}] 自然语言大幅超出，LLM摘要",
                estimated_savings=count_tokens(content) - count_tokens(text)
            ))

        else:
            # 15%≤超出<40% 或结构化内容 → 滑动窗口（首尾保留）
            text = self._sliding_window(text, budget)
            decisions.append(CompressionDecision(
                method=CompressionMethod.SLIDING,
                overflow_ratio=overlap_ratio := overflow_ratio,
                content_type=content_type,
                reason=(
                    f"[{section_name}] "
                    + ("结构化内容禁止LLM摘要，" if content_type in ("code","json","log") else "")
                    + "滑动窗口保留首尾"
                ),
                estimated_savings=count_tokens(content) - count_tokens(text)
            ))

        return text

    # ── 具体压缩方法 ─────────────────────────────────────────────────────────

    def _truncate(self, text: str, budget: int) -> str:
        """尾部截断，保留前 budget 个 tokens 对应的行"""
        lines = text.splitlines(keepends=True)
        result, used = [], 0
        for line in lines:
            lt = count_tokens(line)
            if used + lt > budget:
                break
            result.append(line)
            used += lt
        return "".join(result)

    def _sliding_window(self, text: str, budget: int) -> str:
        """滑动窗口：保留首部 + 省略标记 + 尾部
        
        首部 = budget * 0.6
        尾部 = budget * 0.4 - overlap
        中间插入 '...[已省略 N tokens]...'
        """
        head_budget = int(budget * 0.6)
        tail_budget = int(budget * 0.4) - self.sliding_window_overlap

        lines = text.splitlines(keepends=True)
        # 收集首部
        head_lines, used = [], 0
        for line in lines:
            lt = count_tokens(line)
            if used + lt > head_budget:
                break
            head_lines.append(line)
            used += lt

        # 收集尾部（从末尾往前）
        tail_lines, used = [], 0
        for line in reversed(lines):
            lt = count_tokens(line)
            if used + lt > tail_budget:
                break
            tail_lines.insert(0, line)
            used += lt

        omitted_tokens = count_tokens(text) - count_tokens("".join(head_lines)) - count_tokens("".join(tail_lines))
        omit_marker = f"\n... [已省略约 {omitted_tokens} tokens 的中间内容] ...\n"

        return "".join(head_lines) + omit_marker + "".join(tail_lines)

    def _dedup_lines(self, text: str) -> str:
        """去除完全重复的行（保留首次出现）"""
        seen, result = set(), []
        for line in text.splitlines(keepends=True):
            normalized = line.strip()
            if normalized not in seen:
                seen.add(normalized)
                result.append(line)
        return "".join(result)

    def _llm_summarize(self, text: str, budget: int, section_name: str) -> str:
        """调用 LLM 对内容做高保真摘要
        
        若 LLM 调用失败，自动降级为滑动窗口。
        """
        prompt = (
            f"请将以下「{section_name}」章节内容压缩为不超过 {budget} tokens 的摘要，"
            f"要求：\n"
            f"1. 保留所有关键结论、数据和决策依据\n"
            f"2. 删除重复、冗余、过渡性表述\n"
            f"3. 保持原有逻辑结构\n"
            f"4. 使用中文输出\n\n"
            f"原文：\n{text}"
        )
        try:
            summary = self.llm.complete(prompt, max_tokens=self.llm_summary_max_tokens)
            # 验证摘要未超预算（LLM 有时会超出）
            if count_tokens(summary) > budget:
                summary = self._truncate(summary, budget)
            return summary
        except Exception as e:
            print(f"⚠️ LLM 摘要失败 ({e})，降级为滑动窗口")
            return self._sliding_window(text, budget)


# ═══════════════════════════════════════════════════════════════════════════════
# ContextBuilder（整合 HybridCompressor）
# ═══════════════════════════════════════════════════════════════════════════════

class ContextBuilder:
    """上下文构建器 - GSSC流水线"""

    def __init__(
        self,
        memory_tool: Optional[MemoryTool] = None,
        rag_tool: Optional[RAGTool] = None,
        config: Optional[ContextConfig] = None,
        llm=None   # 传入后 HybridCompressor 可启用 LLM 摘要
    ):
        self.memory_tool = memory_tool
        self.rag_tool = rag_tool
        self.config = config or ContextConfig()
        self._encoding = tiktoken.get_encoding("cl100k_base")
        self.last_quality_report: Optional[ContextQualityReport] = None

        # ← 新增：混合压缩器
        self._compressor = HybridCompressor(
            llm=llm,
            llm_summary_min_overflow=0.40,
            llm_summary_max_tokens=500,
            sliding_window_overlap=50,
            enable_dedup=True
        )
        self.last_compression_result: Optional[CompressionResult] = None

    # ...existing code...

    def _compress(self, context: str) -> str:
        """Compress: 混合压缩（替换原有简单截断）"""
        if not self.config.enable_compression:
            return context

        budget = self.config.get_available_tokens()
        result = self._compressor.compress(context, budget)
        self.last_compression_result = result  # 供外部检查

        if result.method_used != CompressionMethod.SKIP:
            print(f"🗜️  {result}")
            for d in result.decisions:
                print(f"   ↳ [{d.method}] {d.reason} | 节省 {d.estimated_savings} tokens")

        return result.compressed_text

    def print_compression_report(self) -> None:
        """打印最近一次压缩报告"""
        if self.last_compression_result is None:
            print("⚠️  尚未执行压缩")
        else:
            r = self.last_compression_result
            print(f"\n{'─'*50}")
            print(f"🗜️  压缩报告")
            print(f"   方法链      : {r.method_used}")
            print(f"   原始 tokens : {r.original_tokens}")
            print(f"   压缩后      : {r.final_tokens}")
            print(f"   压缩率      : {r.compression_ratio:.2%}")
            for d in r.decisions:
                print(f"   ↳ [{d.method:12s}] {d.reason}")
            print(f"{'─'*50}\n")

    # ...existing code...


# ...existing code...
````

### Q2：Docker 部署架构是怎么设计的？

**A：**

```yaml
# docker-compose.yml
services:
  backend:
    build: ./backend          # Conda 环境 (hello-agents 兼容性)
    ports: ["8000:8000"]
    env_file: ./backend/.env
    volumes: [./backend/logs:/app/logs]   # 日志持久化
    healthcheck:
      test: curl -f http://localhost:8000/
      interval: 30s
      start_period: 90s       # Conda 环境初始化较慢
  
  frontend:
    build:
      context: ./frontend
      args:
        - VITE_API_BASE_URL=http://backend:8000    # 编译期注入
        - VITE_AMAP_WEB_JS_KEY=${VITE_AMAP_WEB_JS_KEY}
    ports: ["80:80"]
    depends_on:
      backend: { condition: service_healthy }

networks:
  trip-network: { driver: bridge }
```

**关键设计决策**：
1. **后端用 Conda 而非 pip**：hello-agents 框架有复杂依赖链，Conda 环境隔离更可靠
2. **前端多阶段构建**：Node 20 构建 → Nginx alpine 运行，镜像从 ~1GB 压缩到 ~30MB
3. **Nginx SSE 透传**：`proxy_buffering off` + `proxy_cache off` + `chunked_transfer_encoding on`
4. **健康检查 start_period=90s**：给 Conda 环境初始化足够时间，避免容器被误判为不健康
